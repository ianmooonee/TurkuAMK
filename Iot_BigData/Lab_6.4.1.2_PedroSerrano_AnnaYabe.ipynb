#### Step 2: Capture an image using the Camera module
###### a) Grab an image, and display it.
We will use your laptop webcam to capture an image.

```python
def takePicture():
    cap = cv2.VideoCapture(0) # video capture source camera (Here webcam of laptop) 
    ret,frame = cap.read() # return a single frame in variable `frame`
    cap.release() # Close video capturing
    return frame
```

```python
def displayPicture(frame):
        cv2.imshow('image',frame) #display the captured image
        cv2.waitKey(0)   # Wait time in milliseconds if 0, press any key to continue
        cv2.destroyAllWindows()
```

```python
# Test these functions
image1 = takePicture()# ... take picture
displayPicture(image1) # Preview "image" in new window
```

## Part 2: Capture images with Python
In this part of the lab, you will capture an image and display it.

###### Step 2: Capture an image.
The code below captures an image and displays it. First, any existing `camera` is closed to ensure we are starting from the beginning with the camera initialization. Then, the camera image properties are set on the `camera` object. Then, the image is captured with the `camera.capture()` method. Finally, the graphic is displayed with `matplotlib.pyplot`. 

```python
image = takePicture()
#cv2.imwrite("test.jpg", image) # Will save image on disk

pltFrame = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) ## Color space conversion from BRG to RGB Without this colors are mixed

#The plt.show() method is used as if the image were a graph.

# Turn off the axis
plt.axis('off')

# Title of the picture
plt.title("Webcam Video/Image")

# show the frame
plt.imshow(pltFrame)
plt.show()
```
